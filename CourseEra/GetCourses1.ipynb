```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def extract_title_from_card(course_card):
    """
    Extracts the title of a course from a course card element.
    """
    try:
        title_element = course_card.find_element(By.CSS_SELECTOR, '[id^="cds-react-aria-"][id$="-product-card-title"]')
        return title_element.text
    except Exception as e:
        print(f"Error extracting title: {e}")
        return None

def go_to_next_page(page_number):
    """
    Constructs the URL for the next page and navigates to it.
    """
    try:
        # Construct the next page URL directly using f-strings
        next_page_url = f"https://www.coursera.org/courses?query=artificial%20intelligence&page={page_number}"
        
        # Navigate to the next page
        driver.get(next_page_url)
        
        # Wait for the page to load (or specific element)
        WebDriverWait(driver, 5).until(EC.presence_of_element_located(
            (By.CSS_SELECTOR, '[id^="cds-react-aria-"][id$="-product-card-title"]')
        ))
        return next_page_url
    except Exception as e:
        print(f"Error navigating to the next page: {e}")
        return None

# Set up Chrome WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode
driver = webdriver.Chrome(options=options)

try:
    # Starting URL and page number
    page_number = 120
    base_url = f"https://www.coursera.org/courses?query=artificial%20intelligence&page={page_number}"
    driver.get(base_url)

    while page_number <= 130:
        print(f"Scraping Page: {page_number}")
        print(f"Current URL: {driver.current_url}")
        
        try:
            # Extract all course cards
            course_cards = WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((
                By.CSS_SELECTOR, "#main > div.css-q9p10k > div > div > div.cds-9.css-0.cds-11.cds-grid-item.cds-56 > div > div.cds-9.css-1xdhyk6.cds-11.cds-grid-item.cds-56.cds-81 > div > div > ul > li > div > div"
            )))

            # Iterate through course cards to extract titles
            for index, course_card in enumerate(course_cards, start=1):
                course_title = extract_title_from_card(course_card)
                if course_title:
                    print(f"Course Title {index}: {course_title}")

        except Exception as e:
            print(f"Error scraping page {page_number}: {e}")
        
        # Navigate to the next page
        page_number += 1
        next_page_url = go_to_next_page(page_number)
        if next_page_url:
            print(f"Navigated to Page: {page_number}, URL: {next_page_url}")
        else:
            print("Failed to navigate to the next page.")
            break

except Exception as e:
    print("Error:", e)
finally:
    driver. Quit() 
```
